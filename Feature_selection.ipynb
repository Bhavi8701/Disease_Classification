```python
def correlation(df, threshold):
    col_corr = set()
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(x_train,0.9)
len(set(corr_features))
```

# Feature selection -Variance threshold

This feature selector removes all low-variance features

```python
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(x_train)
```

```python
sum(var_thres.get_support())
##True means all that features have non-zero variance
#False - columns have to be removed
```

```python
var_thres.get_support()
```

```python
constant_columns = [column for column in x_train.columns
                   if column not in x_train.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
for feature in constant_columns:
    print(feature)
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
Scaler.fit(x_train)

#transform train and test sets
x_train_scaled = Scaler.transform(x_train)
x_test_scaled = Scaler.transform(x_test)
```

```python
Scaler.mean_
```

# Effect of scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.scatter(x_train['LRRFIP1'], x_train['RGS6'])
ax1.set_title("Before Scaling")
ax2.scatter(x_train_scaled[:,0], x_train_scaled[:,2], color ='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

#before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['LRRFIP1'],ax=ax1)
sns.kdeplot(x_train['RGS6'],ax=ax1)

#After scaling
ax2.set_title("After Scaling")
sns.kdeplot(x_train_scaled[:,0], ax=ax2) 
sns.kdeplot(x_train_scaled[:,2], ax=ax2)
```

# Feature selection - using Information gain

```python
from sklearn.feature_selection import mutual_info_classif
mutual_info = mutual_info_classif(x_train,y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index=x_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(20,8))
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
sel_best_cols = SelectKBest(mutual_info_classif, k=20)
sel_best_cols.fit(x_train.fillna(0),y_train)
x_train.columns[sel_best_cols.get_support()]
```
